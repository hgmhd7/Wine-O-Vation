```python
import pandas as pd
```

```python
wine_table_df = pd.read_csv("master_wine_table.csv")

wine_table_df.head()
```

```python
wine_table_df.shape
```

```python
wine_type_list = []

for index, row in wine_table_df.iterrows():
    
    if row['red'] == 1:
        wine_type_list.append('Red')
        
    elif row['rose'] == 1:
        wine_type_list.append('Rose')
        
    elif row['white'] == 1:
        wine_type_list.append('White')
        
    elif row['sparkling'] == 1:
        wine_type_list.append('Sparkling')
        
        
print(wine_type_list[:10], len(wine_type_list))
```

```python
wine_table_df['wine_type'] = wine_type_list

wine_table_df.head(3)
```
